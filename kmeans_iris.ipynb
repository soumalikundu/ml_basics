```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=50)
```

### n_init

It specifies the number of times the algorithm will run with different initial centroid seeds.

```python
y_kmeans = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:,2], X_scaled[:,3], c=y_kmeans, cmap='viridis', marker='o', edgecolors='k', alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], s=300, c='red', marker='X', label="Centroids")
plt.title("K-Means Clustering on Iris Dataset (Petal Features)")
plt.xlabel("Standardized Petal Length")
plt.ylabel("Standardized Petal Width")
plt.legend()
plt.show()
```
